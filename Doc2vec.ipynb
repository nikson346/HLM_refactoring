```python
if datatype<2:
    file_info=EMdata.read_relion(file_path)
    if datatype==0:
        #read data (relion3.1)
        dataset=file_info.getRdata_31()
        optics=file_info.extractoptic()
    else:
        #read relion 3.0
        dataset=file_info.getRdata()
    metadata=dataset[0]
    print(metadata)
    data=dataset[1]
    print(data[0])
    corpus_information=EMdata.process_helical(dataset).extarct_helical()
else:
    #read cryosparc
    dataset=np.load(file_path)
    corpus_information=EMdata.process_cryosparc_helical(dataset).extract_helical()
corpus_dic=corpus_information[0]
corpus=list(corpus_dic.values())
corpus_backup=corpus[:]
```

```python
def read_corpus(corpus):
    for i, tokens in enumerate(corpus):
        yield gensim.models.doc2vec.TaggedDocument(tokens, [i])
train_corpus=list(read_corpus(corpus))
```

```python
model = gensim.models.doc2vec.Doc2Vec(vector_size=100, min_count=2, epochs=40)
```

```python
model.build_vocab(train_corpus)
```

```python
model.train(train_corpus, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
vector = model.infer_vector(train_corpus[1].words)
filament_list=[]
for i in range(len(train_corpus)):
    filament_list.append(model.docvecs[i])

print(len(filament_list))
```
